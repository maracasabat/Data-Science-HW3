```python
import matplotlib.pyplot as plt
import random
```

```python
X = [0]
Y = [0]

def next_pt(x, y, num):
    if num == 0:
        X.append(x / 2.0)
        Y.append(y / 2.0)
    elif num == 1:
        X.append((x + 1) / 2.0)
        Y.append(y / 2.0)
    else:
        X.append((x + 0.5) / 2.0)
        Y.append((y + (3**0.5)) / 2.0)
                   
                   
for _ in range(1000000):
    n = random.randint(0, 2)
    next_pt( X[-1], Y[-1], n)
```

```python
plt.scatter(X, Y, s=0.001)
plt.rcParams["figure.figsize"] = (15, 12)
plt.title('Sierpinski Triangle')
plt.show()
```
