```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

У таблиці наведено дані про взаємонезалежні доходності трьох акцій: A, B, C.

```python
link = 'https://textbook.edu.goit.global/python/data-science-xf8b2g-pwa/v1/uk/docs/module-03/hw'
data = pd.read_html(link)[0]
data
```

Побудуйте лінії регресії, за допомогою функції regplot, залежності доходності однієї акції від іншої. Усього три графіки. Виберіть пару акцій, які найменш залежні в доходностях від один одного.

```python
data.corr()
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
sns.heatmap(data.corr(), annot=True, cmap='BrBG');
```

```python

plt.rcParams["figure.figsize"] = (16, 10)
sns.regplot(x=data['Доходность A, %'], y=data['Доходность B, %'], data=data);
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
sns.regplot(x=data['Доходность A, %'], y=data['Доходность C, %'], data=data);
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
sns.regplot(x=data['Доходность B, %'], y=data['Доходность C, %'], data=data);
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
sns.regplot(x=data['Доходность A, %'], y=data['Доходность B, %'], data=data, label = 'A vs B')
sns.regplot(x=data['Доходность A, %'], y=data['Доходность C, %'], data=data, label = 'A vs C')
sns.regplot(x=data['Доходность B, %'], y=data['Доходность C, %'], data=data, label = 'B vs C')
plt.legend();
```
