Створіть нову таблицю python_data, в якій будуть тільки рядки з фахівцями, які вказали мову програмування Python

```python
python_data = df[ df["Язык.программирования"] == "Python"]
python_data
```

Використовуючи метод groupby, виконайте груповання по стовпцю "Посада"

```python
py_d = python_data.groupby('Должность')
py_d.head()
```

Створіть новий DataFrame, де для згрупованих даних по стовпцю "Посада", виконайте агрегацію даних за допомогою методу agg та знайдіть мінімальне та максимальне значення у стовпці "Зарплата.в.месяц"

```python
df4 = python_data.groupby('Должность').agg({'Зарплата.в.месяц':['min', 'max']})
df4
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(salary):
  return salary.mean().round(2)
```

```python
avg = df.groupby('Должность').agg('Зарплата.в.месяц').apply(fill_avg_salary)
df4['avg'] = avg
df4
```

Створіть описову статистику за допомогою методу describe нового стовпця

```python
df4['avg'].describe()
```

Збережіть отриману таблицю в файл CSV

```python
df4.to_csv('python_data.csv', encoding='utf8')
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
df4.plot(kind='barh', label=['average', 'min', 'max'])
plt.xlabel("Зарплата", fontsize=15, color="midnightblue")
plt.ylabel("Посада", fontsize=15, color="midnightblue")
plt.title("Розподіл заробітньої плати по посадам", fontsize=20, color='r')
plt.legend()
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
sns.barplot(x='Уровень.английского', y='Зарплата.в.месяц', hue='Должность', data=python_data)
plt.xlabel('Рівень англійської', fontsize=15, color="midnightblue")
plt.ylabel('Заробітна плата', fontsize=15, color="midnightblue")
plt.title("Розподіл заробітньої плати від  рівня англійської мови", fontsize=20, color='r')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
sns.countplot(x='Должность', data=python_data)
plt.xlabel('Посада', fontsize=15, color="midnightblue")
plt.ylabel('Кількість працівників', fontsize=15, color="midnightblue")
plt.title("Кількість працівників за посадами", fontsize=20, color='r')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (16, 10)
sns.swarmplot(x='Общий.опыт.работы', y='Зарплата.в.месяц', hue='Должность', data=python_data)
plt.xlabel('Досвід роботи (загальний)', fontsize=15, color="midnightblue")
plt.ylabel('Заробітна плата', fontsize=15, color="midnightblue")
plt.title("Розподіл заробітньої плати в залежності від досвіду роботи", fontsize=20, color='r')
plt.tick_params(labelrotation = 5)
plt.show()
```
